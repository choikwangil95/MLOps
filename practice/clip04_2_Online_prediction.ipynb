### 0. 필요한 Library Import

```python
from google.cloud import aiplatform

import numpy as np
from PIL import Image
```

### 1. Endpoint 설정

```python
# projects/651078960027/locations/us-central1/endpoints/395771409441226752

endpoint = aiplatform.Endpoint(
    endpoint_name="projects/651078960027/locations/southamerica-west1/endpoints/395771409441226752")
```

### 2. Online Inference 이미지 가져오기 & numpy로 변환

```python
IMAGE_PATH = "./practice_image/test_image.jpg"
im = Image.open(IMAGE_PATH)
resized_img = im.resize((80,80))
```

```python
resized_img
```

```python
x_test = np.asarray(resized_img).astype(np.float32)[np.newaxis, :, :, :].tolist()
```

### 3. 추론 완료

```python
endpoint.predict(instances=x_test).predictions
```
